```python
#binomial distribution probability density function
#for flipping a coin, heads or tails (success: heads)
#explanation of math: http://www.stat.yale.edu/Courses/1997-98/101/binom.htm
#code: Robert Silver 2021

import numpy as np
from matplotlib import pyplot as plt
from scipy.special import binom
```

```python
font = {
    'color': 'lightgreen', 'size': 12
}

def getBinomProb(k, n, p):
    return binom(n, k)*(p**k)*((1-p)**(n-k))
```

```python
#number of dice rolls
n = 20;
#probability each coin flip being heads (or dice rolls being 6)
#for dice: p=.16667
p = .166667;
#x axis
intArr = np.arange(0, 21, 1);
probList = [];
for val in intArr:
    probList.append(getBinomProb(val, n, p))
probArr = np.array(probList)

#it doesn't matter which side we're getting probability for, all sides same p
plt.title("Binomial distro: dice rolls being  side 3 out of 20 total", fontdict=font );
plt.xlabel("number of rolls landing on side", fontdict=font)
plt.ylabel("percentage likelihood this many rolls landing", fontdict=font)
plt.bar(intArr, probArr, width=0.8, align='center')
plt.show()
```

```python
#number of dice rolls
n = 20;
#probability each coin flip being heads (or dice rolls being 6)
#for dice: p=.16667
p = .5;
#x axis
intArr = np.arange(0, 21, 1);
probList = [];
for val in intArr:
    probList.append(getBinomProb(val, n, p))
probArr = np.array(probList)

plt.title("Binomial distro: coin flips landing on heads, 20 flips", fontdict=font);
plt.xlabel("number of flips on heads", fontdict=font)
plt.ylabel("percentage likelihood this many heads", fontdict=font)
plt.bar(intArr, probArr, width=0.8, align='center')
plt.show()




```
